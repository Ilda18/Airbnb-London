# Data Cleaning

#Reviews_per_month, number_of_reviews and number_of_reviews_ltm. are similar metric. I'll keep only the last one.

```python
df.drop(['Neighbourhood_group'], 
        axis=1, inplace=True)
```

```python
# Also, I will change the type of data for last-review from object to datetime.
df['last_review'] = pd.to_datetime(df['last_review'])
df
```

```python
# change data types
df['last_review'] = pd.to_datetime(df['last_review'], format = '%d-%m-%Y' )
df.info()
```

```python
# sorting data by Order Date
df.sort_values (by = ['last_review'], inplace = True, ascending = True)
```

#Checking Null Data

```python
#1. dealing with null values
nulls = df.isnull().sum()
nulls = pd.DataFrame(nulls[nulls > 0]).reset_index()
nulls = nulls.rename(columns={'index': 'column_name', 0: 'number_of_nulls'})
rows_number = df.shape[0]
nulls['null_percentage'] = round((nulls['number_of_nulls'] / rows_number) *100, 2)
nulls = nulls.sort_values(by=['null_percentage'], ascending=1).reset_index().iloc[:, 1:] 
nulls
```

```python
#Rename columns on necesary coulmns
df.rename(columns = {'id' : 'Id'}, inplace = True)
df.rename(columns = {'minimum_nights' : 'Min_nights'}, inplace = True)
df.rename(columns = {'neighbourhood' : 'Neighbourhood '}, inplace = True)
df.rename(columns = {'room_type' : 'Room_type'}, inplace = True)
df.rename(columns = {'last_review' : 'Last_review'}, inplace = True)
df.rename(columns = {'availability_365' : 'Availability_365'}, inplace = True)
df.rename(columns = {'name' : 'Name'}, inplace = True)
df.rename(columns = {'host_id' : 'Host_id'}, inplace = True)
df.rename(columns = {'price' : 'price'}, inplace = True)
df.rename(columns = {'price' : 'price'}, inplace = True)
df.rename(columns = {'calculated_host_listings_count' : 'Calculated_listing_count'}, inplace = True)
df.rename(columns = {'number_of_reviews_ltm' : 'Nr_of_reviews_ltm'}, inplace = True)
df
```
